```python
import warnings
warnings.filterwarnings('ignore')
```

```python
income=223
spend=150
```

```python
import pickle
file1=open('scale.pkl','rb')
scale=pickle.load(file1)
file2=open('model.pkl','rb')
model=pickle.load(file2)
file3=open('plot.pkl','rb')
plot=pickle.load(file3)
file4=open('centroid.pkl','rb')
centroid=pickle.load(file4)
```

```python
import numpy as np
a=[income,spend]
x=np.array([a])
x=scale.transform(x)
Y_pred=model.predict(x)[0]
if Y_pred ==0:
    print ('focus/target')
elif Y_pred==1:
    print ('sensitive')
elif Y_pred==2:
    print ('ignore')
elif Y_pred==3:
    print ('miser')
else:
    print ('careless')
```

```python
#Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(data=plot[0],x='INCOME',y='SPEND',color='red',label='focus/target')
sns.scatterplot(data=plot[1],x='INCOME',y='SPEND',c='blue',label='sensitive/ignore')
sns.scatterplot(data=plot[2],x='INCOME',y='SPEND',c='green',label='ignore')
sns.scatterplot(data=plot[3],x='INCOME',y='SPEND',c='orange',label='Miser')
sns.scatterplot(data=plot[4],x='INCOME',y='SPEND',c='yellow',label='Careless')
#to show centroid of each categories
sns.scatterplot(data=centroid,x=centroid[:,0],y=centroid[:,1],c='purple',marker='*',s=200)
sns.scatterplot(x=[a[0]],y=[a[1]],c='black',marker='^',s=300,label='customer')
plt.title('Classification of customer')
plt.show()
```
